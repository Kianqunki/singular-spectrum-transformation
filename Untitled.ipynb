```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from glob import glob
from fastsst import SingularSpectrumTransformation
```

## ECG

```python
ecgs = glob("all_data/ECG_data/*.txt")
```

```python
ecg_datasets = []
for ecg in ecgs:
    df = pd.read_csv(ecg,header=None,delimiter="\t")
    ecg_datasets.append(df.values)
```

```python
plt.figure(figsize=(20,5))
plt.plot(ecg_datasets[0][:,1:])
```

```python
x = ecg_datasets[0][:,1]
sst = SingularSpectrumTransformation(500)
score = sst.score_offline(x)

f,ax = plt.subplots(2,1,figsize=(20,10))
ax[0].plot(x);
ax[1].plot(score,"r")
```
